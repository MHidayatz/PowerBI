## PowerBi Theme

If we have a repeatative background image to use for every page in the PowerBI report, we can use a json theme to make it easier to manage.


1. Go to https://www.base64-image.de/ to convert an image to Base64.
2. Once uploaded we will need to click on copy image to  save the URL.

image.png

3. Create a json file with IDE of your preference, I am using notepad to write the following json code:

```
JSON File:
{"name":"MyTheme"

,"visualStyles":{"page":{"*":{"background":
    [
     {
                        "image": {
                            "name": "MyBackground",
                            "scaling": "Fill",
                            "url": "Enter Base64 Image Here"
                        },
                        "transparency": 0
                    }
    ]
   }
  }
 }
}

```

4. Replace "My Background" with the link to the Image.
5. Open PowerBI file and select browse for themes.




image.png

6. Open the JSON file and you should see the image imported as a theme background image for every page.
